```python
# Map credibility labels to binary values
credibility_map = {
    "low credibility": 1,
    "medium credibility": 1,
    "high credibility": 0
}
df["credibility_score"] = df["credibility_label"].map(credibility_map)

# Add domain dummies to the DataFrame
# Create dummy variables for the domain
domain_dummies = pd.get_dummies(df['domain_type'], prefix='domain')

# Concatenate the dummy variables with the original DataFrame
df = pd.concat([df, domain_dummies], axis=1)

# Calculate correlation of all numeric features with credibility_score
correlation_with_credibility = df.corr(numeric_only=True)["credibility_score"] \
    .drop("credibility_score") \
    .sort_values(ascending=False)

# Show top correlated features
print(correlation_with_credibility)
```

```python
# Drop the original label column
df = df.drop(columns=["credibility_label"])
```

```python

# Identify all non-numeric (categorical) columns
categorical_cols = df.select_dtypes(exclude='number').columns

# One-hot encode all categorical columns
df_encoded = pd.get_dummies(df, columns=categorical_cols, prefix=categorical_cols, drop_first=True)

# Compute correlation with credibility_score
correlation_with_credibility = df_encoded.corr(numeric_only=True)["credibility_score"] \
    .drop("credibility_score") \
    .sort_values(ascending=False)

# Show top correlations

correlation_with_credibility
```

```python

# Select top 20 features by absolute correlation
top_corr = correlation_with_credibility.abs().sort_values(ascending=False).head(30)
top_corr_signed = correlation_with_credibility.loc[top_corr.index]  # Keep original sign

# Now sort by signed value so positive are on top, negative on bottom
top_corr_signed = top_corr_signed.sort_values(ascending=False)

# Plot the top correlations
plt.figure(figsize=(10, 6))
sns.barplot(x=top_corr_signed.values, y=top_corr_signed.index)
plt.xlabel("Correlation with Credibility Score")
#plt.title("Top 25 Feature Correlations with Credibility Score")
plt.tight_layout()
plt.show()
```

```python

# Assume correlation_with_credibility is your Series of feature correlations
N = 25
top_corr = correlation_with_credibility.abs().sort_values(ascending=False).head(N)
top_corr_signed = correlation_with_credibility.loc[top_corr.index].sort_values(ascending=False)

# Light color palette
colors = ['lightblue' if val > 0 else 'lightcoral' for val in top_corr_signed]

# Plot
plt.figure(figsize=(10, 6))
bars = sns.barplot(x=top_corr_signed.values, y=top_corr_signed.index, palette=colors)

# Add labels inside bars
for value, bar in zip(top_corr_signed.values, bars.patches):
    plt.text(
        x=bar.get_width()/1.3,                       # center of bar
        y=bar.get_y() + bar.get_height() / 2,        # center vertically
        s=f"{value:.3f}",                             # formatted correlation
        ha='center', va='center', color='black', fontsize=8
    )

# Labels and layout
plt.xlabel("Correlation with Credibility Score")
#plt.title("Top 20 Feature Correlations with Credibility Score")
plt.tight_layout()
plt.show()

```

## 📊 Interpretation: Feature Correlation with Credibility Score

This chart shows how different HTML and CSS features correlate with `credibility_score` (0 = low, 1 = medium, 2 = high).

### ✅ Positive Correlations (Associated with Higher Credibility)

- **Title Length:** Websites with longer `<title>` tags tend to score higher in credibility. This may indicate more structured, descriptive metadata.
- **HTML Size:** Larger HTML files are often found in more professional websites that contain rich, structured content.
- **Number of `<iframe>`, `<div>`, and CSS Selectors:** These structural elements suggest complex layout and modular design, often seen in reputable media sources.

### ❌ Negative Correlations (Associated with Lower Credibility)

- **Text-to-HTML Ratio:** Lower ratios (i.e., lots of HTML and little text) are common in low-credibility websites, potentially indicating cluttered, style-heavy layouts with little meaningful content.
- **Text Length:** Shorter text content is associated with less credible sites, reinforcing the idea that content-rich pages are more trustworthy.
- **Animations, Transitions, and Excess Stylesheets:** Heavy use of animations and transitions may correlate with visually aggressive or distracting interfaces, often found in spammy or sensationalist websites.

### 🔍 Other Observations

- Features like `font-weight: bold`, number of images, and number of colors show weak correlation, suggesting that **visual emphasis alone does not strongly indicate credibility**.
- Slightly negative correlations for `!important` flags and complex selectors could reflect **less clean or maintainable CSS**, potentially tied to less reputable sources.

### 🧠 Summary

Credibility is positively associated with **structured, content-rich websites** and negatively associated with **clutter, excessive styling, and minimal textual content**. These insights can inform both feature engineering and interpretation for modeling or classification tasks.
